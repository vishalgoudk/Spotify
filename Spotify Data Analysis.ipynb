```python
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity',ascending = False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[15]
```

```python
# Converting msec in to seconds

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace= True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

# To ignore all warnings
warnings.filterwarnings("ignore")

# Assuming df_tracks is your DataFrame

# Drop non-numeric columns
non_numeric_columns = ["key", "mode", "explicit"]
numeric_df = df_tracks.drop(non_numeric_columns, axis=1)

# Calculate correlation matrix
corr_df = numeric_df.corr(method="pearson")

# Create a new figure
plt.figure(figsize=(15, 7))

# Create a heatmap
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")

# Set the title of the plot
plt.title("Correlation HeatMap Between Variables")

# Rotate the x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))

print(len(sample_df))
```

```python
plt.figure(figsize= (15,7))
sns.regplot(data= sample_df, y = "loudness", x= "energy", color= "skyblue").set(title= "Loudness VS Energy Correlation")
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="orange")
plt.title("Popularity VS Acousticness Correlation")
plt.show()
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist")
plt.title("Number of songs per year")
plt.show()
```

```python
total_dr = df_tracks['duration']
sns.set_style(style="whitegrid")
fig_dims = (15, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.lineplot(x=years, y=total_dr, ax=ax)
plt.title("Year vs Duration")
plt.xticks(rotation=60)
plt.show()
```
